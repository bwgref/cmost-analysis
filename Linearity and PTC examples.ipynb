```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from cmost_image import get_file_list

font = {'size' : 14, 'family' : 'sans-serif'}
matplotlib.rc('font', **font)

cmost_dir = './local_data'
```

```python
# Load data from the 001 data set
all_frames, ro_modes, dates, exp_times, voltages, temps, signals, rms = get_file_list(cmost_dir,file_prefix='cmost001',debug=True)
```

```python
# Plot the linearity curve
plt.figure(figsize=[9,5])
plt.loglog()
plt.title('Linearity @ {}V'.format(voltages[0]))
plt.plot(exp_times,signals,'o-')
plt.xlabel('Exposure Time (ms)')
plt.ylabel('Signal (adu)')
plt.show()
plt.close()
```

```python
# Plot the photon transfer curve
plt.figure(figsize=[9,5])
plt.loglog()
plt.title('PTC @ {}V'.format(voltages[0]))
plt.plot(signals[rms>0],rms[rms>0],'o-')
# Plot shot noise level
plt.plot(signals,10**(0.5*np.log10(signals))/2)
plt.xlabel('Signal (adu)')
plt.ylabel('rms (adu)')
plt.show()
plt.close()
```
