```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from cmost_image import open_cmost_file, get_corrected_level

font = {'size' : 14, 'family' : 'sans-serif'}
matplotlib.rc('font', **font)

cmost_dir = './local_data'
```

```python
# Display cmost001_20201007031546.fits raw frame
frames, ro_mode, date, exp_time, voltage, temp = open_cmost_file('{}/cmost001_20201007031546.fits'.format(cmost_dir))    
print('Exposure time: {} ms\nReadout mode: {}\nLED voltage: {} V\nDate: {}'.format(exp_time,ro_mode,voltage,date))  

plt.figure(figsize=[16,5])
plt.imshow(frames[0])
plt.colorbar(fraction=0.05, aspect=15, shrink=0.75)
plt.show()
plt.close()
```

```python
# Subtract reset from the signal to get actual recorded image
corrected_level = get_corrected_level(frames[0])

plt.figure(figsize=[9,5])
plt.imshow(corrected_level)
plt.colorbar(fraction=0.05, aspect=15, shrink=0.85)
plt.show()
plt.close()
```
